```python
from requests import get
```

```python
response = get("https://api.spacexdata.com/v3/launches")
```

```python
response_json = response.json()
```

```python
response_json[0]
```

```python
import pandas as pd
```

```python
space_x_all_launches = pd.DataFrame(response_json)
```

```python
space_x_all_launches.iloc[0]
```

```python
space_x_all_launches.info()
```

```python
space_x_all_launches.iloc[0]["launch_site"]["site_name"]
```

```python
space_X_cleaned_data = space_x_all_launches.drop(["last_wiki_update", "launch_date_source", "last_wiki_revision","last_wiki_launch_date","last_ll_update", "last_ll_launch_date","crew"], axis=1)
```

```python
space_X_cleaned_data.info()
```

```python
import numpy as np
```

```python
space_X_cleaned_data["launch_success"]=space_X_cleaned_data["launch_success"].fillna(False)
```

```python
space_X_cleaned_data["launch_success"]
```

```python
space_X_cleaned_data["launch_window"].value_counts()
```

```python
space_X_cleaned_data["launch_window"]=space_X_cleaned_data["launch_window"].fillna(0)
```

```python
space_X_cleaned_data.drop("launch_failure_details",axis=1, inplace=True)
```

```python
space_X_cleaned_data["details"]=space_X_cleaned_data["details"].fillna("Empty")
```

```python
space_X_cleaned_data.info()
```

```python
space_X_cleaned_data["static_fire_date_unix"]=space_X_cleaned_data["static_fire_date_unix"].fillna(0)
```

```python
space_X_cleaned_data["static_fire_date_utc"]=space_X_cleaned_data["static_fire_date_utc"].fillna(0)
```

```python
space_X_cleaned_data=space_X_cleaned_data.drop(["timeline", "last_date_update"], axis=1)
```

```python
space_X_cleaned_data.info()
```

```python
space_X_cleaned_data.launch_year=space_X_cleaned_data.launch_year.astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=space_X_cleaned_data, x='launch_success')
plt.title("Launch Success Distribution")
plt.show()
```

```python
sns.countplot(data=space_X_cleaned_data,hue="launch_success", x=space_X_cleaned_data["launch_site"].apply(lambda x: x["site_name"]))
```

```python
sns.countplot(data=space_X_cleaned_data,hue="launch_success", x=space_X_cleaned_data["rocket"].apply(lambda x: x["rocket_name"]))
```

```python
 space_X_cleaned_data.corr(numeric_only=True)
```

```python
space_X_cleaned_data.mission_id=space_X_cleaned_data.mission_id.apply(lambda x: " ".join(x))
```

```python
space_X_cleaned_data.rocket=space_X_cleaned_data.rocket.apply(lambda x: f"{x["rocket_id"]} {x["rocket_name"]}")
```

```python
space_X_cleaned_data.launch_site=space_X_cleaned_data.launch_site.apply(lambda x: f"{x["site_id"]} {x["site_name"]}")
```

```python
space_X_cleaned_data.links=space_X_cleaned_data.links.apply(lambda x: x["mission_patch"])
```

```python
space_X_cleaned_data.ships=space_X_cleaned_data.ships.apply(lambda x: " ".join(x))
```

```python
space_X_cleaned_data.drop("telemetry", axis=1, inplace=True)
```

```python
pd.get_dummies(data=space_X_cleaned_data)
```

```python
X=pd.get_dummies(data=space_X_cleaned_data.drop(["launch_success", "links", "mission_name"], axis=1))
y=space_X_cleaned_data.launch_success
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=321)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_train = lr.predict(X_train)
```

```python
y_pred_train
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = lr.predict(X_test)
```

```python
r2_score(y_test, y_pred_test)
```
